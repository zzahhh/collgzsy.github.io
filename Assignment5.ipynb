## 读取数据

```python
import pandas as pd
data = pd.read_csv('data/wine_quality/winequality-white.csv', delimiter=";")
```

```python
# YOUR CODE HERE
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict

features = ['fixed acidity']
x1 = data[features]
y1 = data['quality']
features = ['fixed acidity','volatile acidity']
x2 = data[features]
y2 = data['quality']
features = ['fixed acidity','citric acid']
x3 = data[features]
y3 = np.log(data['quality'])

model1=LinearRegression()
model2=LogisticRegression()
model3=LinearDiscriminantAnalysis()
predict1 = cross_val_predict(model1,x1,y1,cv = 10)
predict2 = cross_val_predict(model1,x2,y2,cv = 10)
predict3 = np.exp(cross_val_predict(model1,x3,y3,cv = 10))
predictm2 = cross_val_predict(model2,x1,y1,cv = 10)
predictm3 = cross_val_predict(model3,x1,y1,cv = 10)
print(accuracy_score(predict1.astype(np.int64),y1))
print(accuracy_score(predict2.astype(np.int64),y2))
print(accuracy_score(predict3.astype(np.int64),y1))
print(accuracy_score(predictm2.astype(np.int64),y1))
print(accuracy_score(predictm3.astype(np.int64),y1))



```
